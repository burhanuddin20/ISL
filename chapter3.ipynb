The cubic model is more flexible, and so is likely to overfit the training data meaning that the fit of the model will be affected by variance in the training data that is not representive of the true f(x).

Since the true relationship is linear, the additional terms in the cubic regression model will fit the noise in the training data, rather than the true underlying relationship.
This overfitting leads to a lower training RSS, but it reduces the model's ability to generalize to new test data, as the extra terms capture noise rather than signal.

The linear model has a higher bias because it assumes a simpler (linear) relationship, but it also has lower variance, as it does not try to fit the noise in the data.
The cubic model has lower bias because it can represent more complex relationships, but it has higher variance, as it is more sensitive to fluctuations in the training data.
