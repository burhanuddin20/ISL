### Derivation of $( R^2 = \text{Cor}(x, y)^2 )$ in Simple Linear Regression

#### Assumptions:
1. Mean of \(x\) and \(y\) is zero: $ \bar{x} = \bar{y} = 0\ $.
2. Simple linear regression model: $(y_i = \beta x_i + \epsilon_i).$

---

#### Definition of $(R^2)$:
$[
R^2 = \frac{\text{TSS} - \text{RSS}}{\text{TSS}} = 1 - \frac{\text{RSS}}{\text{TSS}} =  \frac{\text{ESS}}{\text{TSS}},
]$
where:
-  $ (\text{TSS} = \sum_{i=1}^n y_i^2) $ (Total Sum of Squares),
- $(\text{RSS} = \sum_{i=1}^n (y_i - \hat{y}_i)^2)$ (Residual Sum of Squares),
- $(\text{ESS} = \sum_{i=1}^n \hat{y}_i^2)$ (Explained Sum of Squares).$


Given that $( \hat{y}_i = \beta x_i )$, we can write:
$[
\text{ESS} = \sum_{i=1}^n (\hat{y}_i)^2 = \beta^2 \sum_{i=1}^n x_i^2.
]$


---

#### Estimate for $(\beta)$:
$[
\beta = \frac{\sum_{i=1}^n x_i y_i}{\sum_{i=1}^n x_i^2}.
]$

Substituting $(\beta)$ into $(\text{ESS})$:
$[
\text{ESS} = \left(\frac{\sum_{i=1}^n x_i y_i}{\sum_{i=1}^n x_i^2}\right)^2 \sum_{i=1}^n x_i^2 = \frac{(\sum_{i=1}^n x_i y_i)^2}{\sum_{i=1}^n x_i^2}.
]$

---

#### TSS:
$[
\text{TSS} = \sum_{i=1}^n y_i^2.
]$

---

#### $(R^2)$:
$[
R^2 = \frac{\text{ESS}}{\text{TSS}} = \frac{\frac{(\sum_{i=1}^n x_i y_i)^2}{\sum_{i=1}^n x_i^2}}{\sum_{i=1}^n y_i^2}.
]$

Simplify:
$[
R^2 = \frac{(\sum_{i=1}^n x_i y_i)^2}{\sum_{i=1}^n x_i^2 \cdot \sum_{i=1}^n y_i^2}.
]$

---

#### Correlation $((\text{Cor}(x, y)))$:
The correlation between \(x\) and \(y\) is:
$[
\text{Cor}(x, y) = \frac{\sum_{i=1}^n x_i y_i}{\sqrt{\sum_{i=1}^n x_i^2 \cdot \sum_{i=1}^n y_i^2}}.
]$

Square it:
$[
\text{Cor}(x, y)^2 = \frac{(\sum_{i=1}^n x_i y_i)^2}{\sum_{i=1}^n x_i^2 \cdot \sum_{i=1}^n y_i^2}.
]$

---

#### Conclusion:
$[
R^2 = \text{Cor}(x, y)^2.
]$
