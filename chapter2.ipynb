Are any of the predictors associated with per capita crime rate?
If so, explain the relationship.

Crime rate can be correlated with LSTAT (higher crime in lower-status areas) and with DIS (perhaps areas closer to city centers). You may see that CRIM is highly skewed, with a few neighborhoods having very high values.

Do any of the suburbs of Boston appear to have particularly
high crime rates? Tax rates? Pupil-teacher ratios? Comment on
the range of each predictor.

```python
print("Range (min, max, and range) for each predictor:\n")
for col in boston.columns:
    col_min = boston[col].min()
    col_max = boston[col].max()
    col_range = col_max - col_min
    print(f"{col}: min={col_min}, max={col_max}, range={col_range}")
    
# 5. Identify particularly high CRIM (crime rate), TAX (tax rate), PTRATIO (pupil-teacher ratio).
#    Thresholds below are somewhat arbitrary; feel free to adjust.

# Crime rate threshold - let's pick CRIM > 20 as "high"
high_crim = boston[boston['crim'] > 20]
print("\nSuburbs with particularly high crime rates (CRIM > 20):")
print(high_crim)

# Tax rate threshold - let's pick TAX > 600 as "high"
high_tax = boston[boston['tax'] > 600]
print("\nSuburbs with particularly high tax rates (TAX > 600):")
print(high_tax)

# PTRATIO threshold - let's pick PTRATIO > 20 as "high"
high_ptratio = boston[boston['ptratio'] > 20]
print("\nSuburbs with particularly high pupil-teacher ratios (PTRATIO > 20):")
print(high_ptratio)
```
